### Ejercicio 4 — Alerta de temperaturas

Temperaturas registradas en 10 días:
`[22, 35, 18, 41, 29, 38, 17, 33, 26, 40]`

Imprime:
- El número de día (desde 1) y la temperatura de cada día que superó 30°C
- Cuántos días superaron ese umbral
- `"Alerta calor"` si fueron más de 3 días

````python
temps = [22, 35, 18, 41, 29, 38, 17, 33, 26, 40]
dia = 0
count =0
# Tu solución:
for i in temps:
    dia+=1
    if i >= 30:
        print(f"Dia: {dia} Obtuvo una temperatura de {i}°C")
        count+=1

if count > 2:
    print(f"Alerta de calor, superamos los {count} dias de calor")
````

### Ejercicio 5 — Inventario con alertas

```python
inventario = {
    "manzanas": 50,
    "peras": 12,
    "uvas": 0,
    "mangos": 34,
    "fresas": 7
}
```

Recorre el diccionario e imprime:
- Productos con stock entre 1 y 10 → `"Stock bajo"`
- Productos con stock 0 → `"Agotado"`
- Al final, el total de unidades disponibles en toda la tienda

````python
inventario = {
    "manzanas": [50,30,20],
    "peras": 12,
    "uvas": 0,
    "mangos": 34,
    "fresas": 7
}

# 1. Inicializamos el acumulador para el total
total_unidades = 0

# 2. Recorremos el diccionario usando .items()
for producto, stock in inventario.items():
    # Sumamos al total en cada vuelta
    total_unidades += stock
    
    # Lógica de alertas
    if stock == 0:
        print(f"{producto}: Agotado")
    elif stock > 1 and stock < 10:
        print(f"{producto}: Stock bajo ({stock} unidades)")

print(f"Total de unidades en la tienda: {total_unidades}")
````

### Ejercicio 7 — Adivina el número

Número secreto guardado en una variable: `secreto = 42`

El usuario tiene intentos ilimitados para adivinarlo.
Después de cada intento incorrecto indica si debe subir o bajar:
- `"Muy bajo, sube."`
- `"Muy alto, baja."`

Cuando adivine, imprime cuántos intentos necesitó.

````python
secreto = 42
intentos = 3
num = 0

while num != 42 and intentos > 0:
    num = int(input(f"Adivina el número secreto ({intentos} intentos restantes): "))
    
    if num > 42:
        print("Muy alto, baja")
        intentos -= 1
    elif num < 42:
        print("Muy bajo, sube")
        intentos -= 1

if intentos == 0 and num != 42:    
    print("Lo siento, fallaste los intentos. Era el", secreto)
else:
    print("¡Felicidades, adivinaste el número secreto!")
````

### Ejercicio 8 — Validador de contraseña

Pide que el usuario cree una contraseña que cumpla estas dos reglas:
- Mínimo 8 caracteres
- Que no sea solo números (`cadena.isdigit()` devuelve `True` si todo son dígitos)

Mientras no cumpla las dos reglas, indica cuál está fallando y pide de nuevo.
Cuando sea válida: `"Contraseña registrada correctamente."`

````python
# Tu solución:
print("Hola creea una contraseña segura")

while True:
    contraseña = input("Ingrese su contraseña: ")

    if len(contraseña) < 8:
        print("Contraseña débil: debe tener al menos 8 caracteres, Vuelve a intentarlo")
    else:
        if not any(char.isalpha() for char in contraseña):
            print("Contraseña débil: debe contener al menos una letra, Vuelve a intentarlo")
        else: 
            print(f"La Contraseña: {contraseña} segura y registrada")
            break
````
