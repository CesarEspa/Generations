## Paso 2 — Las operaciones

Reemplaza el `pass` del Paso 1 con el código que:

1. Calcula el resultado según la operación elegida
2. Maneja el caso especial de **división por cero** — si `num2` es 0, muestra un mensaje de error y no operes
3. Guarda cada operación en el historial con este formato:

```
5.0 + 3.0 = 8.0
```

> - Para agregar a la lista: `historial.append("texto")`  
> - Para armar el texto: `f"{num1} + {num2} = {resultado}"`

Cuando termines, prueba las cuatro operaciones y luego elige la opción 5 para ver el historial.

````python
# Copia aquí tu calculadora completa con las operaciones implementadas:

historial = []

print("===========================")
print("   CALCULADORA INTERACTIVA ")
print("===========================")

while True:
    print("\n¿Qué quieres hacer?")
    print("1. Sumar")
    print("2. Restar")
    print("3. Multiplicar")
    print("4. Dividir")
    print("5. Ver historial")
    print("6. Salir")

    opcion = input("\nElige una opción (1-6): ")

    if opcion == "6":
        print("¡Hasta luego!")
        break

    elif opcion == "5":
        if len(historial) == 0:
            print("El historial está vacío.")
        else:
            print("\n--- Historial de operaciones ---")
            for i, operacion in enumerate(historial, 1):
                print(f"{i}. {operacion}")

    elif opcion in ["1", "2", "3", "4"]:
        num1 = float(input("Primer número: "))
        num2 = float(input("Segundo número: "))

        # ── Tu código continúa desde acá (Paso 2) ──
        if opcion == "1":
            resultado = num1 + num2
            print("La suma es de: ", resultado)
            historial.append(f"{num1} + {num2} = {resultado}")
        elif opcion == "2":
            resultado = num1 - num2
            print("La resta es de: ", resultado)
            historial.append(f"{num1} - {num2} = {resultado}")
            
        elif opcion == "3":
            resultado = num1 * num2
            print("La Multiplicacion es de: ", resultado)
            historial.append(f"{num1} * {num2} = {resultado}")
        else:
            if num2 != 0:
                resultado = num1 / num2
                print("La division es: ", resultado)
                historial.append(f"{num1} / {num2} = {resultado}")
            else:
                print("No se puede divir entre 0")

    else:
        print("Opción no válida. Elige un número del 1 al 6.")
````
