Ayer construiste una calculadora que funciona. Hoy la vas a mejorar.

No vas a reescribirla desde cero — vas a **reorganizarla**. Ese proceso tiene nombre: **refactorización**.

## Paso 1 — Analiza antes de tocar

Lee este código. Es una versión simplificada de tu calculadora de ayer.  
**No lo ejecutes aún.** Primero responde las preguntas de abajo.

```python
# Calculadora original — léela completa antes de continuar

historial = []

while True:
    print("\n1. Sumar  2. Restar  3. Multiplicar  4. Dividir  5. Historial  6. Salir")
    opcion = input("Opción: ")

    if opcion == "6":
        print(f"Operaciones realizadas: {len(historial)}")
        break

    elif opcion == "5":
        if len(historial) == 0:
            print("Historial vacío.")
        else:
            for i, op in enumerate(historial, 1):
                print(f"{i}. {op}")

    elif opcion in ["1", "2", "3", "4"]:
        num1 = float(input("Primer número: "))
        num2 = float(input("Segundo número: "))

        if opcion == "1":
            resultado = num1 + num2
            simbolo = "+"
        elif opcion == "2":
            resultado = num1 - num2
            simbolo = "-"
        elif opcion == "3":
            resultado = num1 * num2
            simbolo = "*"
        elif opcion == "4":
            if num2 == 0:
                print("Error: no se puede dividir entre cero.")
                continue
            resultado = num1 / num2
            simbolo = "/"

        registro = f"{num1} {simbolo} {num2} = {resultado}"
        historial.append(registro)
        print(f"Resultado: {resultado}")

    else:
        print("Opción no válida.")
```

## Paso 2 — Las operaciones

El bloque que calcula el resultado según la operación elegida es el candidato más obvio.  
Está en el centro del programa, tiene lógica propia, y si mañana agregas una quinta operación, solo tocarías esa función.

### Tu tarea

Crea una función llamada `calcular` que:
- Reciba tres parámetros: `num1`, `num2` y `opcion`
- Aplique la operación correcta con `if/elif`
- Maneje la división por cero internamente
- **Devuelva** el resultado y el símbolo de la operación

> Python puede devolver dos valores separados por coma:  
> `return resultado, simbolo`  
> Y recibirlos así: `resultado, simbolo = calcular(5, 3, "1")`

Prueba tu función con al menos tres llamadas antes de continuar.

```python
# Tu código acá:

def calcular(num1, num2, operacion):
    if operacion == "1":
        resultado = num1 + num2
        simbolo = "+"
    elif operacion == "2":
        resultado = num1 - num2
        simbolo = "-"
    elif operacion == "3":
        resultado = num1 * num2
        simbolo = "*"
    elif operacion == "4":
        if num2 == 0:
            return "Error: no se puede dividir entre cero."
        resultado = num1 / num2
        simbolo = "/"
    else:
        return "Operación no válida."
    return resultado, simbolo



# Pruebas — descomenta y ejecuta:
print(calcular(10, 5, "1"))   # debería dar (15.0, '+')
print(calcular(10, 5, "4"))   # debería dar (2.0, '/')
print(calcular(10, 0, "4"))   # debería manejar la división por cero
```

## Paso 4 — Mostrar el historial

El bloque que muestra el historial tiene una responsabilidad clara y delimitada.  
Es exactamente el tipo de código que se beneficia de tener nombre propio.

### Tu tarea

Crea una función llamada `mostrar_historial` que:
- Reciba la lista `historial` como parámetro
- Imprima cada operación numerada
- Si el historial está vacío, muestre un mensaje apropiado

> Pregunta para reflexionar: ¿esta función necesita `return`?

```python
def mostrar_historial(historial):
    # Tu código acá:
    if len(historial) == 0:
       print("Historial vacío.")
    else:
       for i, op in enumerate(historial, 1):
         print(f"{i}. {op}")



# Prueba:
mostrar_historial([])
mostrar_historial(["5.0 + 3.0 = 8.0", "10.0 / 2.0 = 5.0"])
```
